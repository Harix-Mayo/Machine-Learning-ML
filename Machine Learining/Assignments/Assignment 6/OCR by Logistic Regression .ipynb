```python
import numpy as np

# Loading data into numpy 

Train_X = np.loadtxt('trainX.txt', dtype=np.uint8)
Col_Of_Ones = np.ones((Train_X.shape[0]), dtype=np.uint8)      # Create col of ones for training data set
Train_X = np.insert(Train_X , 0 , Col_Of_Ones , axis = 1)      # Append col of ones in training data 

Train_Y = np.loadtxt('trainY.txt', dtype=np.uint8)
Train_Y = np.where(Train_Y == 2 , 1 , 0)                       # Mapping of 2 and 4 class to 1 and 0 respectively

Test_X = np.loadtxt('testX.txt', dtype=np.uint8)
Col_Of_Ones = np.ones((Test_X.shape[0]), dtype=np.uint8)       # Create col of ones for testing data set
Test_X = np.insert(Test_X , 0 , Col_Of_Ones , axis = 1)        # Append col of ones in testing data 

Test_Y = np.loadtxt('testY.txt', dtype=np.uint8)
Test_Y = np.where(Test_Y == 2 , 1 , 0)                         # Mapping of 2 and 4 class to 1 and 0 respectively

Etas = [0.01 , 0.1 , 0.3 , 0.5]                                # List of etas
for Eta in Etas:
    
    Thetas = np.random.uniform(0, 1, size=(Train_X.shape[1]))      # Randomly initialize weights
    iterations = 1
    m = Train_X.shape[0]
    Predictions = np.where(sigmoid(np.dot(Train_X , Thetas)) >= 0.5 , 1 , 0)        # Predictions on training data
    Accuracy = calculate_Accuracy(Predictions , Train_Y)

    while (not(Accuracy == 100) and (iterations <= 1000)):

        Probabilities = sigmoid(np.dot(Train_X , Thetas))                           # Calculate Probabilities by Sigmoid function
        Predictions = np.where(Probabilities >= 0.5 , 1 , 0)                        # Assign class after thresholding      
        Thetas = Thetas - (Eta / m) * np.dot(Train_X.T , (Probabilities - Train_Y)) # Update weights 
        Accuracy = calculate_Accuracy(Predictions , Train_Y)
        print ('Accuracy for iteration: ' , iterations , ' is: ' , Accuracy)
        #print ('Cost on Training Data: ' , (1/Train_X.shape[0]) * ((-1 * np.dot(Train_Y.T, np.log(Probabilities) / np.log(10))) - (np.dot((1 - Train_Y).T , np.log(1 - Probabilities) / np.log(10)))))
        iterations += 1


    print ('\nFor Eta: ' , Eta)
    print ('Total no. of Iterations: ' , iterations - 1)
    Predictions = np.where(sigmoid(np.dot(Train_X , Thetas)) >= 0.5 , 1 , 0)
    print ('Accuracy on Training data: ' , calculate_Accuracy(Predictions , Train_Y))
    Predictions = np.where(sigmoid(np.dot(Test_X , Thetas)) >= 0.5 , 1 , 0)
    print ('Accuracy on Testing data: ' , calculate_Accuracy(Predictions , Test_Y) , '\n')
 
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# To calculate accuracy
def calculate_Accuracy(Predictions , Actuals):
    return np.where(Predictions == Actuals , 1 , 0).sum() / Actuals.shape[0] * 100
```
